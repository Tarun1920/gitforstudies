A t-test and a z-test are both statistical methods used to assess whether the means of two groups are significantly different from each other. However, they differ in their application and assumptions.

T-Test:
Use Case:

Scenario: You have a small sample size (typically less than 30) and want to compare the means of two groups.
Example: You are comparing the average scores of two groups of students (Group A and Group B) after a different teaching method is applied to each group. Each group consists of 20 students.
Assumption:

Assumes that the population follows a normal distribution.
More suitable for small sample sizes.
Formula:

The t-test statistic is calculated as the difference between the sample means divided by the standard error.
Z-Test:
Use Case:

Scenario: You have a large sample size (typically greater than 30) and want to compare the means of two groups.
Example: You are comparing the average heights of two groups of people (Group X and Group Y), where each group has more than 30 individuals.
Assumption:

Assumes that the population follows a normal distribution.
More suitable for large sample sizes.
Formula:

The z-test statistic is calculated similarly to the t-test but uses the population standard deviation instead of the sample standard deviation.
Summary:
Use a t-test when dealing with a small sample size or when the population standard deviation is unknown.
Use a z-test when dealing with a large sample size and when the population standard deviation is known.

One-Tailed Test:
A one-tailed test, also known as a one-sided test, is used when you want to test a specific hypothesis about the direction of the effect or difference in your data. It is appropriate when you have a clear expectation of the direction in which the effect will occur.

Two-Tailed Test:
A two-tailed test, also known as a two-sided test, is used when you want to test a hypothesis without specifying the direction of the effect. It is typically used when you are interested in detecting whether there is any significant difference, without assuming whether it's an increase or decrease.

Type I Error (False Positive):
A Type I error occurs when you reject a null hypothesis that is actually true. In other words, it's a false positive or a "false alarm." The significance level (alpha, typically set at 0.05) of a test represents the probability of making a Type I error. Lowering the alpha level reduces the risk of Type I errors but increases the risk of Type II errors.

Example Scenario:
Suppose a pharmaceutical company is testing a new drug's effectiveness in reducing blood pressure. The null hypothesis (H0) is that the drug has no effect on blood pressure. However, due to random sampling variability, the test erroneously concludes that the drug significantly lowers blood pressure when, in reality, it doesn't.

Type II Error (False Negative):
A Type II error occurs when you fail to reject a null hypothesis that is actually false. It's a false negative or a "missed opportunity." The probability of making a Type II error is influenced by the power of a statistical test, which depends on factors like sample size and effect size. Increasing the sample size or the effect size generally reduces the risk of Type II errors but may increase the risk of Type I errors.

Example Scenario:
Consider a quality control process in a manufacturing plant where the null hypothesis (H0) is that the production process is within acceptable limits. However, the process is producing defective products. The test fails to detect this issue, and as a result, the defective products continue to be produced and shipped to customers.

Summary:
Type I Error (False Positive): Rejecting a true null hypothesis.
Type II Error (False Negative): Failing to reject a false null hypothesis.

A point estimate (e.g., sample mean or proportion) for the parameter of interest.
